# EDA of Breast cancer dataset

## 1. Loading the dataset

```python
import pandas as pd
df = pd.read_csv('Breast Cancer dataset.csv')
df.head()
```

## 2. understanding the dataset

```python
print("Rows :",df.shape[0])
print("Columns :",df.shape[1])
df.info()
```

```python
df.columns
```

```python
df.describe()
```

## 3. Missing value analysis

```python
total=df.isnull().sum().sort_values(ascending=False)
percent=(df.isnull().sum()/len(df)*100).sort_values(ascending=False)
missing_data=pd.concat( [total,percent], axis=1,keys=["Total Missing","Percentage"])
missing_data.head(10)
```

### Observation

The dataset contains only one column ('Unnamed: 32') with missing values. This column has 569 missing entries, which means 100% of the values are missing. Since the column contains no useful information, it can be safely removed from the dataset.

## 4. Data Cleaning

```python
df.drop(columns=["Unnamed: 32"],inplace=True,errors="ignore")
print(df.shape)
print(df.isnull().sum().sum())
```

## 5. Duplicate Records

```python
df.duplicated().sum()
```

## 6. Largest and Smallest tumors

```python
largest=df.nlargest(5,"area_mean")
print("Largest\n")
print(largest[["diagnosis","radius_mean","area_mean"]])

print("\nSmallest\n")
smallest=df.nsmallest(5,"area_mean")
print(smallest[["diagnosis","radius_mean","area_mean"]])
```

## 6. Correlation with Diagnosis

```python
df["diagnosis_num"]=df["diagnosis"].map({"B":0,"M":1})
```

```python
corr=df.corr(numeric_only=True)["diagnosis_num"].sort_values(ascending=False)
print(corr.head(10))
```

## 7. Diagnosis Distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="diagnosis", data=df)
plt.title("Distribution of Diagnosis")
plt.show()
```

```python
diag=df["diagnosis"].value_counts()
print(diag)
print("\nPercentage:")
print((diag/len(df)*100).round(2))
```

## 8. Histogram of radius_mean

```python
df["radius_mean"].hist(bins=20)
plt.title("Distribution of Radius Mean")
plt.xlabel("Radius Mean")
plt.ylabel("Frequency")
plt.show()
```

## 9. Boxplot of radius_mean

```python
sns.boxplot(y=df["radius_mean"])
plt.title("Boxplot of Radius Mean")
plt.show()
```

### avg radius by diagnosis

```python
print(df.groupby("diagnosis")["radius_mean"].mean().round(2))
```

## 10.Correlation Heatmap

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

## 11. Pair Plot

```python
sns.pairplot(df,vars=["radius_mean","texture_mean","perimeter_mean","area_mean"],hue="diagnosis")
plt.show()
```

## 12. Histogram of texture_mean

```python
df["texture_mean"].hist(bins=20)
plt.title("Distribution of Texture Mean")
plt.xlabel("Texture Mean")
plt.ylabel("Frequency")
plt.show()
```

## 13. Boxplot of texture_mean

```python
sns.boxplot(y=df["texture_mean"])
plt.title("Boxplot of Texture Mean")
plt.show()
```

## 14. diagnosis vs radius box plot

```python
sns.boxplot(x="diagnosis",y="radius_mean",data=df)
plt.title("Diagnosis vs Radius Mean")
plt.show()
```

## 15. Outlier Detection

```python
Q1=df["radius_mean"].quantile(0.25)
Q3=df["radius_mean"].quantile(0.75)
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
df[(df["radius_mean"]<lower)|(df["radius_mean"]>upper)]
```

## 16. Skewness & Kurtosis

```python
print(df["radius_mean"].skew())
print(df["radius_mean"].kurt())
```
